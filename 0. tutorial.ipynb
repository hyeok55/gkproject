```python
# 실습문제 풀기 전 실행해 주세요.
df1 = {'key' : ['a','b','c','d'],
       'A' : [1,2,3,4]}
df2 = {'key' : ['a','b','c','d'],
       'B' : [5,6,7,8]}
df1 =  pandas.DataFrame(df1)
df2 = pandas.DataFrame(df2)
```

```python
# 실습문제2 소스코드
pandas.merge(df1,df2,on="key")

```
